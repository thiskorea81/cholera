```python
import numpy as np
import pandas as pd
import folium
from scipy.spatial.distance import cdist
```

```python
펌프위치 = pd.read_csv('/content/cholera/펌프 위치.csv')
콜레라환자 = pd.read_csv('/content/cholera/콜레라 환자.csv')
```

```python
펌프위치
```

```python
펌프위치 = 펌프위치[['경도', '위도']].to_numpy()
```

```python
펌프와의거리 = cdist(콜레라환자[['경도', '위도']].values, 펌프위치)
가까운펌프번호 = np.argmin(펌프와의거리, axis=1)
```

```python
colors = ['red', 'yellow', 'purple', 'deeppink', 'lime', 'aqua', 'blue', 'orange']
지도 = folium.Map(location=[51.5149, -0.1348], zoom_start=16)
```

```python
for 펌프번호, row in 콜레라환자.iterrows():
    folium.CircleMarker(location=(row['경도'], row['위도']),radius=5,
                        color=colors[가까운펌프번호[펌프번호]], fill=True).add_to(지도)
```

```python
for center in 펌프위치:
    folium.Marker(location=center,icon=folium.Icon(color='blue', icon='star')).add_to(지도)
```
